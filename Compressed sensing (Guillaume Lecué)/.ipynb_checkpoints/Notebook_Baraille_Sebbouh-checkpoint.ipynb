## 0. Some preliminary tools

### (a) Discrete Fourier Transforms and Discrete Fourier Transforms in 1D and 2D

```python
def fftc(x):
    """1D Discrete Fourier Transform"""
    return 1 / np.sqrt(x.shape[0]) * np.fft.fftshift(np.fft.fft(np.fft.ifftshift(x))) 

def ifftc(x):
    """1D Inverse Discrete Fourier Transform"""
    return np.sqrt(x.shape[0]) * np.fft.fftshift(np.fft.ifft(np.fft.ifftshift(x))) 
```

```python
def fft2c(x):
    """2D Discrete Fourier Transform"""
    return 1 / np.sqrt(x.shape[0] * x.shape[1]) * np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(x))) 

def ifft2c(x):
    """2D Inverse Discrete Fourier Transform"""
    return np.sqrt(x.shape[0] * x.shape[1]) * np.fft.fftshift(np.fft.ifft2(np.fft.ifftshift(x)))
```

### (b) Computing similarities between images

We use two measures of similarity ($l_2$ and SSIM) to compare the reconstructed images with the original one. The SSIM has been developed to measure the visual quality of a compressed image, compared to the original image. The idea of the SSIM is to compare the structure of two images, rather than a pixel-to-pixel difference. We also use the $l_2$ difference.

```python
def similarities(image_original, image):
    """Function that computes the SSIM and the l2 similarity between two images.
       Two similar images have a SSIM close to 1 and a small l2 similarity"""
    
    img_original = np.abs(image_original)
    img = np.abs(image)
    
    SSIM = ssim(img_original, img)
    l2_diff = np.linalg.norm(np.asarray(img_original) - np.asarray(img),2)
    l2_sim = l2_diff / np.linalg.norm(np.asarray(img_original),2)
    
    return SSIM, l2_sim
```

### (c) Displaying the k-space

```python
def kspace(data):
    """Fucntion taht displays the k-space"""
    image = np.abs(fft2c(data))
    image = (image - image.min())/(image.max())
    image_mod = image.copy()
    
    # Modifying the extreme values
    min_val = 1e-5
    changes_min = np.sum(image < min_val) / 512**2 *100
    idx_min = np.where(image < min_val)
    image_mod[idx_min] = min_val
    max_val = 0.01
    changes_max = np.sum(image > max_val) / 512**2 *100
    idx_max = np.where(image > max_val)
    image_mod[idx_max] = max_val
    
    changes = changes_min + changes_max
    print('Modifying {0}% of the data'.format(changes))
    plt.imshow(1-image_mod, cmap='Greys')
```

## 1. Heuristics on Sparse Signals

Here we give some heuristics in 1D that will help us to understand how to revover the signal from a 2D image.

### (a) Sparse signals

We will attempt to denoise a sparse signal that is corrupted by random noise. Consider the model 
$$
Y = x + \epsilon
$$
where $x$ is sparse and $\epsilon$ is a random noise. We want to find an estimate $\hat{x}$ of $x$.

```python
def signal(n, s, sigma=0.05):
    """Construct a noisy signal of lenght n with s gaussian variables localized randomly and zeros everywhere else"""
    x = np.zeros(n)
    for i in range(s):
        x[i] = (i+1) / s
    random.shuffle(x)
    x_true = x.copy()
    return x_true, x + np.random.normal(0, sigma, n)
```

```python
n = 128
s = 5
x, y = signal(n, s)
```

```python
plt.figure(figsize=(15, 5))
plt.stem(y)
plt.xlabel("Coefficients")
plt.ylabel("Magnitude")
plt.title("Coefficients of noisy y")
```

### (b) Recovering sparse signals unsing the $\mathcal{l}^1$ norm

We now want to show the sparsifyig properties of the $\mathcal{l}_1$ norm. We solve:
$$ 
argmin \frac{1}{2} \| x - y \|_2^2 + \lambda \frac{1}{2} \| x \|_1
$$
The solution to this problem is known to be:
$$
 \begin{equation}
    \hat{x} =  prox_{\lambda |.|}(y) =
    \begin{cases}
      y + \lambda & \text{if}\ y < - \lambda \\
      0 & \text{if}\ - \lambda < y < \lambda \\
      y - \lambda & \text{if}\ y > \lambda \\
    \end{cases}
  \end{equation}
$$

```python
def SoftThresh(y, lbd):
    """Soft Thresholding Function"""
    return np.sign(y) * np.maximum(0, np.absolute(y)-lbd)
```

```python
lbd = 0.2
x_hat1 = SoftThresh(y, lbd)
```

```python
plt.figure(figsize=(15, 5))
plt.stem(x_hat1)
plt.xlabel("Coefficients")
plt.ylabel("Magnitude")
plt.title("Estimated coefficients of y using the l1 norm")
```

```python
# Perfect support recovery
np.where(x != 0), np.where(x_hat1 != 0)
```

We get perfect support recovery, but the $l_1$ norm shrinks the coefficients. 

### (c) The importance of incoherent sampling

We explore here the connection between compressed sensing and the importance of incoherent sampling. In the following, we will use capitals to show that variables are in the Fourier space. For example, $X$ is the discrete Fourier transform of $x$.

```python
# x in the Fourier domain
X = fftc(x)
```

In compressed sensing, we undersample the measurements. We measure a subset of the k-space, $X_m =\Gamma_m x$ where $\Gamma_m$ is a Fourier transform evaluated only at a subset of frequency domain samples. This is an underdetermined data set for which there is an infinite number of possible signals. However, we do know that the original signal is sparse, so there is hope we will be able to reconstruct it.


The theory of compressed sensing suggests random undersampling, that enforces incoherence. To see why, we will look at equispaced undersampling and compare it to random undersampling.

**Equispaced sampling**

As showed in the report, one cannot recover the true signal using an equispaced sampling scheme. We illustrate that below:

```python
def equispaced_sampling(x, prop=0.25):
    """Function that samples a proportion prop of coefficients in a coherent way"""
    k = np.int(1/prop)
    X = fftc(x)
    Xu = np.zeros(n, dtype=complex)
    for i in range(n):
        if i % (k) == 0:
            Xu[i] = X[i]
    # Back in the domain of the original x
    xu = ifftc(Xu) * k
    return Xu
```

```python
Xu = equispaced_sampling(x, prop=0.25)
xu = ifftc(Xu)
```

```python
plt.figure(figsize=(15, 5))
plt.stem(np.abs(xu))
plt.xlabel("Coefficients")
plt.ylabel("Magnitude")
plt.title("Coefficients of y with equispaced sampling")
```

We cannont recover the true signal from this undersampling scheme as we obtain replicas of the true signal and we cannot determine which one is the original one. 

**Incoherent sampling**

A way to deal with this issue is to sample in an incoherent way. Random sampling is a way of doing so.

```python
def random_sampling(x, prop=0.25):
    """Function that samples a proportion prop of coefficients in a random way"""
    k = np.int(len(x) * prop)
    X = fftc(x)
    idx = np.random.choice(n, k)
    Xr = np.zeros(n, dtype=complex)
    Xr[idx] = X[idx]
    # Back in the domain of the original x
    xr = ifftc(Xr) * np.int(1/prop)
    return xr
```

```python
xr = random_sampling(x, prop=0.25)
```

```python
plt.figure(figsize=(15, 5))
plt.stem(np.abs(xr))
plt.xlabel("Coefficients")
plt.ylabel("Magnitude")
plt.title("Coefficients of y with random sampling")
```

We now denoise the recovered signal using thresholding.

```python
xr_thresh = SoftThresh(xr, lbd=0.3)

plt.figure(figsize=(10, 4))
plt.stem(np.abs(x))
plt.xlabel("Coefficients")
plt.ylabel("Magnitude")

plt.figure(figsize=(10, 4))
plt.stem(np.abs(xr_thresh))
plt.xlabel("Coefficients")
plt.ylabel("Magnitude")
plt.title(" Denoised coefficients of y with random sampling using thresholding")
```

We can recover the biggest coefficients of the sparse signal using this undersampling scheme.

### (d) Reconstruction from Randomly Sampled Frequency Domain Data

Inspired by the denoising example, we will add an $\mathcal{l}^1$ penalty and solve:
$$ 
argmin \frac{1}{2} \| F_u \hat{x} - y \|_2^2 + \lambda  \| \hat{x} \|_1
$$
In this case, $\hat{x}$ is the estimated signal, $F_u \hat{x}$ is the undersampled Fourier transform of the estimate, and y are the samples of the Fourier transform that we have acquired. 

Now the variables are coupled through the Fourier transform, and there is no closed-form solution. Therefore we will solve it iteratively applying soft-thresholding and constraining data consistency. If $\hat{X} = F\hat{x}$, we initially set $\hat{X}_0 = y$ and implement the following :
- Compute the inverse Fourier transform to get an estimate of the signal $\hat{x}_i = F^* \hat{X}_i$
- Apply SoftThresh $\hat{x}_i = F \hat{x}_i$ in the signal domain
- Compute the Fourier transform $\hat{X}_i = F\hat{x}_i$
- Enforce data consistency in the frequency domain :
$
 \begin{equation}
    \hat{X}_{i+1}[j]=
    \begin{cases}
      \hat{X}_{i}[j] & \text{if}\ y[j]=0 \\
      y[j] & \text{otherwise}
    \end{cases}
  \end{equation}
$
- Repeat until $\|\hat{x}_{i+1} - \hat{x}_i\| < \epsilon$

```python
def SoftThreshComplex(y, lbd, eps=1e-6):
    """Another version of the Soft Thresholding Function taking complex into account"""
    return (np.abs(y) > lbd) * (np.abs(y) - lbd) / (np.abs(y) + eps) * y
```

```python
# Incoherent undersamplig of the signal as described above
xr = random_sampling(x, prop=0.25)
Y = fftc(xr)
```

```python
def POCS(x, Y, lbd=0.1, niter=20):
    """Reconstruction algorithm"""
    history = []
    Xhat = Y
    for i in range(niter):
        xhat = ifftc(Xhat)
        xhat = SoftThreshComplex(xhat, lbd)
        Xhat = fftc(xhat)
        Xhat = Xhat * (Y == 0) + Y
        history.append(max(abs(x - np.absolute(xhat))))
    return history, xhat, np.absolute(xhat)

history, x_pocs, _ = POCS(xr, Y, lbd=0.2)
```

## 2. Sparsity of Medical Images

Medical images can have a sparse representation in a transform domain, such as the wavelet domain.

We use brain.mat which can be downloaded from https://people.eecs.berkeley.edu/~mlustig/CS.html. We compute the Wavelet transform of the image and plot the image using the biggest the wavelet transform coefficients. Wavelet coefficients represent both space and spatial frequency information. Each band of wavelet coefficients represent a scale (frequency band) of the image. The location of the wavelet coefficient within the band represent its location in space. 

```python
image = loadmat('brain.mat')
data = image['im']
plt.imshow(np.abs(data))
plt.title('Original image of the brain')
```

We want to reconstruct the image above from a limited number of samples without reducing too much its quality (we will give measures of similarity with the original image using th complete k-space). More precisely, as this medical image is sparse in the wavelet domain, we want to reconstruct it using a limited number of coefficients in the wavelet domain. 

```python
def sparse_representation(data, k):
    """Representation of the image using only k% of the wavelet coefficients"""
    
    # We use Daubechies wavelets
    w = pywt.Wavelet('db4')
    # In the wavelet domain
    im_W = pywt.dwt2(data, w)
    # Approxiamtion, Horizontal, Vertical and Diagonal coefficients
    cA, (cH, cV, cD) = im_W 
        
    # Sort all coefficients from cA, cH, cV, cD
    all_coefs = np.concatenate((np.abs(cA), np.abs(cH), np.abs(cV), np.abs(cD)), axis=0)
    all_coefs = all_coefs.reshape(1, -1)[0]
            
    # Only retain the top k %
    all_coefs_sorted = np.sort(all_coefs)
    length = len(all_coefs_sorted)
    top_k = int(length * (k / 100))

    thresh = all_coefs_sorted[-top_k:][0]

    cA_sparse = (np.abs(cA) > thresh) * np.abs(cA)
    cH_sparse = (np.abs(cH) > thresh) * np.abs(cH)
    cV_sparse = (np.abs(cV) > thresh) * np.abs(cV)
    cD_sparse = (np.abs(cD) > thresh) * np.abs(cD)
                     
    im_W_sparse = cA_sparse, (cH_sparse, cV_sparse, cD_sparse) 
    # Back in the pixel domain
    im_rec_sparse = pywt.idwt2(im_W_sparse, w)

    return im_rec_sparse
```

```python
ks = [1, 5, 10, 20]

plt.figure(figsize=(15, 5))

for k in range(0, 4):
    plt.subplot(1, 4, k+1)
    plt.imshow(np.abs(sparse_representation(data, ks[k])))
    SSIM, l2_sim = similarities(data, sparse_representation(data, ks[k]))
    plt.title('{0}% : SSIM ={1: .2f} / l2 ={2: .2f}'.format(ks[k], SSIM, l2_sim))
```

In the figure above we reconstruct the brain image using top % coefficients. We then compute the SSIM and the $l_2$ error of each image with the original brain image. Using 20% of the coefficients enables a 'good' reconstruction of the image.

## 3. Reconstruct the image using appropriate masks

We now explore a 2D compressed sensing reconstruction from undersampled data. The goal here is to reconstruct the original brain image using a well-chosen sample of the k-space. We display below the full k-space.

```python
kspace(data)
```

The function below reconstruct an image from a limited number of k-space coefficients using a mask. It applies a 2D Discrete Fourier transform on the image to go into the k-space, it then apply the considered mask on the k-space, and go back to the image space. 

```python
def reconstruction(image, mask):
    """Reconstruts an image by applying a mask"""
    M = fft2c(image)
    M_masked = M * mask 
    im_masked = ifft2c(M_masked)
    return im_masked
```

### (a) The axymptotic incoherence : Completely random masks are not efficient

We showed that sampling the k-space must be done in an incoherent way. This can be done using a completely random sampling. However, incoherence is not enough. An important aspect of frequency-domain sampling is matching the power spectrum of the image. If the energy is concentrated in lower spatial frequencies, more samples should be allocated there. We illustrate that by using two different random masks (which threfore require the incoherent sampling assumption):
- a uniform mask, drawn from a uniform distribution
- a variable density mask, whose density is concentrated in the center of th k-space

The masks are arrays of 1 and 0 of the same shape as the brain image. They are applied on the k-space to subsample it. The masks approximately have the same number of non zeros coefficients. Therefore, we use the same number of samples to reconstruct the image in both cases (33% of coefficients).

Both masks are provdided with the data. We did not build it.

```python
# Approximately the same number of non zero elements
coefs_kspace = data.shape[0] * data.shape[1]
coefs_unif = np.count_nonzero(image['mask_unif'])
coefs_vardens = np.count_nonzero(image['mask_vardens'])

print('Number of non zero elements in the uniform mask:', coefs_unif)
print('Number of non zero elements in the variable density mask:', coefs_vardens)
print('We use in both case {0} % of the coefficients.'.format(np.int(coefs_vardens / coefs_kspace * 100)))
```

** Our own mask using the SDN**

We now build our own SDN mask: we completely sample the center of the k-space and randomly sample the rest of the k-space using a Bernouilli schme.

```python
# Definition of the mask
my_mask = np.random.choice([0, 1], size=(512,512), p=[3./4, 1./4])
for i in range(170, 342):
    for j in range(170, 342):
        my_mask[i, j]=1

# Reconstructing the image using our customized mask        
im_cus = reconstruction(data, my_mask)
```

**Distribution of Lustig**

The article of M. Lustig suggests to subsample the k-space using a distribution function that is decreasing with the distance to the center of the k-space.
$$
P(x, y) = [ 1 - \frac{\sqrt[]{2}}{n}. \sqrt[]{x^2 + y^2} ]^4, \text{ where } x, y \in ]- \frac{n}{2}, \frac{n}{2}]  
$$

```python
def probability(x, y, n):
    return (1 - np.sqrt(2) / n * np.sqrt(x**2 + y**2))**4
```

```python
#Definition of Lustig's mask 

# N-O quarter of the mask
n = 512
n_half = np.int(n/2)
my_mask_smart1 = np.zeros((n_half, n_half))
for x in range(0,n_half):
    for y in range(0, n_half):
        proba = probability(x, y, n)
        my_mask_smart1[x, y] = np.random.binomial(n=1, p=proba)
        
# Three other quarters
my_mask_smart2 = np.round(np.abs(rotate(my_mask_smart1, 90)))
my_mask_smart3 = np.round(np.abs(rotate(my_mask_smart2, 90)))
my_mask_smart4 = np.round(np.abs(rotate(my_mask_smart3, 90)))

# Blending
step1 = np.concatenate((my_mask_smart2, my_mask_smart1), axis=0)
step2 = np.concatenate((my_mask_smart3, my_mask_smart4), axis=0)
my_mask_smart = np.concatenate((step2, step1), axis=1)
```

```python
# Reconstructing the image using Lustig's mask        
im_smart = reconstruction(data, my_mask_smart)
```

### (a) Definition of the algorithms

**Linear reconstruction algorithm**

This is a simple algoithm that samples the k-space accordingly to a mask, as described earlier.

```python
def linear_reconstruction(image, mat_mask, pdf_mask):
    """Function that implements a linear reconstruction algorithm sumbsapling the k-space using a mask"""
    image_masked = image * mat_mask
    im_cs = ifft2c(image_masked / pdf_mask)
    im_cs = ifft2c(fft2c(im_cs) * (1 - mat_mask) + image_masked)
    return im_cs
```

**Thresholding algorithm**

```python
# These utilities functions were inspired by:
# https://inst.eecs.berkeley.edu/~ee123/sp16/hw/hw9_Compressed_Sensing.html

def coeffs2img(LL, coeffs):
    LH, HL, HH = coeffs
    return np.vstack((np.hstack((LL, LH)), np.hstack((HL, HH))))


def unstack_coeffs(Wim):
        L1, L2  = np.hsplit(Wim, 2) 
        LL, HL = np.vsplit(L1, 2)
        LH, HH = np.vsplit(L2, 2)
        return LL, [LH, HL, HH]

    
def img2coeffs(Wim, levels=4):
    LL, c = unstack_coeffs(Wim)
    coeffs = [c]
    for i in range(levels-1):
        LL, c = unstack_coeffs(LL)
        coeffs.insert(0,c)
    coeffs.insert(0, LL)
    return coeffs
    
    
def db_dwt2(im):
    coeffs = pywt.wavedec2(im, wavelet='db4', mode='per', level=4)
    Wim, rest = coeffs[0], coeffs[1:]
    for levels in rest:
        Wim = coeffs2img(Wim, levels)
    return Wim


def db_idwt2(Wim):
    coeffs = img2coeffs(Wim, levels=4)
    return pywt.waverec2(coeffs, wavelet='db4', mode='per')
```

```python
def POCS2(image, mat_mask, pdf_mask, mu, niter, verbose = False):
    image_masked = image * mat_mask
    im_cs = ifft2c(image_masked / pdf_mask)
    w = pywt.Wavelet('db4')
    list_errors = []
    A = np.zeros(image_masked.shape)
    for i in range(niter):
        wave = db_dwt2(im_cs)
        wave = SoftThreshComplex(wave, mu)
        im_cs = db_idwt2(wave)
        im_cs = ifft2c(fft2c(im_cs) * (1 - mat_mask) + image_masked)

        err = np.linalg.norm(fft2c(im_cs) - image, 2) / np.linalg.norm(image, 2)
        list_errors.append(err)
        if i % 20 == 0 and verbose == True:
            print("iteration = {}; error = {}".format(i, err))
            plt.figure
            plt.imshow(np.abs(im_cs), cmap = plt.cm.gray)
            plt.show()
    
    return im_cs
```

**Proximal gradient algorithm**

```python
def spectral_soft_thresh(A, mu):
    U, s, V = svd(A)
    s_mu = (s - mu) * (s > mu)
    return np.dot(U, np.dot(np.diag(s_mu), V))
```

```python
def prox_grad_descent(image, mat_mask, mu, n_iter=100, verbose = False):
    """Function that implements the proximal gradient algorithm"""
    image_masked = np.multiply(image, mat_mask)
    list_errors = []
    A = np.zeros(image_masked.shape)
    for i in range(n_iter + 1):
        A = spectral_soft_thresh(A * (1 - mat_mask) + image_masked, mu)
        err = np.linalg.norm(A - image, 2) / np.linalg.norm(image, 2)
        list_errors.append(err)
        if i % 20 == 0 and verbose == True:
            print("iteration = {}; error = {}".format(i, err))
            plt.figure
            plt.imshow(np.abs(ifft2c(A)), cmap = plt.cm.gray)
            plt.show()
    return ifft2c(A), list_errors
```

**Congugate gradient algorithm**

```python
def f(m, y, l, mask=vardens_mask):
    """computes the objective function"""
    wave = db_dwt2(m)
    #return norme_p(fft2c(m)*mask-y, 2) + l * np.sum(np.abs(wave))
    return np.sum(np.abs(fft2c(m)*mask-y) ** 2) + l * np.sum(np.abs(wave))
```

```python
def gradf(m, y, mu=1e-6, lbd=0.1, mask=vardens_mask, wavelet='haar'):
    """computes the gradient of the objective"""
    
    wt = db_dwt2(m)
    Psi_m = np.diag(np.dot(wt.T, wt))
    Winv = np.zeros(m.shape[0], dtype='complex')
    for i in range(Psi_m.shape[0]):
        root = np.sqrt(Psi_m[i] + mu)
        Winv[i] = 1/root
    Winv = np.diag(Winv)
    grad_pen = db_idwt2(np.dot(Winv, wt))
    
    return 2 * ifft2c(fft2c(m) - y) + lbd * grad_pen 
```

```python
def conj_grad(y=kspace_image, mask=vardens_mask, lbd=0.1, mu=1e-6, wavelet='db', max_iter=30, max_lsiter=100,
           alpha=0.05, beta=0.6, verbose=True):
    '''The conjugate gradient descent algorithm with line search. M.Lustig, Sparse MRI (2007)'''
    m = np.zeros(y.shape, dtype='complex')
    g0 = gradf(m, y, mu, lbd, mask)
    dm = -g0
    t = 1
    for k in range(max_iter):
        
        # line search
        t = 1
        alpha_term = np.real(np.dot(g0.flatten().T, dm.flatten()))
        for i in range(max_lsiter):
            if i % 10 == 0:
                if i < 1:
                    t = t / beta
            if f(m + t * dm, y, lbd, mask) > f(m, y, lbd, mask) + alpha * t * alpha_term:
                t = beta * t
            else:
                break
        
        # cg iteration
        m = m + t * dm
        g1 = gradf(m, y, mu, lbd, mask)
        gamma = np.linalg.norm(g1, 2) / np.linalg.norm(g0, 2)
        dm = -g1 + dm
        g0 = g1
        
        if verbose:
            if k % 1 == 0:
                ssim_k = ssim(np.abs(pixel_image), np.abs(m))
                l2sim_k = np.linalg.norm(np.abs(m-data), 2) / np.linalg.norm(data, 2)
                print('it : {}, g_norm = {}, f = {}'.format(k, np.linalg.norm(g0, 2), f(m, y, lbd, mask)))
                print('ssim = {}, l2 error = {}'.format(ssim_k, l2sim_k))
                
    return m
```
